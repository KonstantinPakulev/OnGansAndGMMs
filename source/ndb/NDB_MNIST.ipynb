# NDB evaluation on MNIST



```
from keras.datasets import mnist
from NDB import NDB

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)

mnist_NDB = NDB(x_train, k_clusters=20)

mnist_NDB.calculate(x_test)
```

```
mnist_NDB.plot_hist([(x_test[:5000], 'VAL_0'), 
                     (x_test, 'VAL_full')])
```

```
mnist_NDB.vizaulize(x_test, (28, 28), mode='min')
```

```
mnist_NDB.vizaulize(x_test, (28, 28), mode='max')
```
