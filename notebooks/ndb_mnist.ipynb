# NDB evaluation on MNIST



```python
%load_ext autoreload
%autoreload 2
```

```python
from keras.datasets import mnist
from source.ndb.ndb import NDB
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
```

```python
mnist_NDB = NDB(x_train, k_clusters=200)

mnist_NDB.calculate(x_test)
```

```python
mnist_NDB.plot_hist([(x_test[:5000], 'VAL_0'), 
                     (x_test, 'VAL_full')])
```

## Visualization of over-occupied bin

```python
mnist_NDB.visualize_min_max_bins(x_test, [28, 28], 1, mode='max')
```

## Visualization of under-occupied bin

```python
mnist_NDB.visualize_min_max_bins(x_test, [28, 28], 1, mode='min')
```
